```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
seq = 10
pth = f"Figs/project_test_K_10_K0_30_seq_{seq}/pred_out.json"
pred_out = json.load(open(pth, "r"))
res_by_rank = {
    k: np.zeros(seq * 3)
    for k, v in pred_out["case_0"].items()
    if k not in ["input_ids", "correct_next_token_ids"]
}
for case in pred_out.values():
    correct_next_token_ids = np.array(case["correct_next_token_ids"])
    for k, v in case.items():
        if k not in ["input_ids", "correct_next_token_ids"]:
            pred = np.array(v["pred_next_token_ids"])
            res_by_rank[k] += np.array(pred == correct_next_token_ids)

d = np.array([v for k, v in res_by_rank.items()])[:, seq * 2 : -1] / 50

for pos in np.arange(0, d.shape[1], step=2):
    plt.plot(d[:, pos])
    plt.title(f"pos{pos}")
    plt.xlabel("Rank")
    plt.show()
```
